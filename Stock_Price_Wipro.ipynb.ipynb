```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv("Wipro.csv")
df
```

```python
import math
median_Deliverable_volume = math.floor(df.Deliverable_volume.median())
median_Deliverable_volume
```

```python
df.Deliverable_volume = df.Deliverable_volume.fillna(median_Deliverable_volume)
df.head()
```

```python
import math
median_percentage_Deliverble = math.floor(df.percentage_Deliverble.median())
median_percentage_Deliverble
```

```python
df.percentage_Deliverble = df.percentage_Deliverble.fillna(median_percentage_Deliverble)
df.head()
```

Predicting TurnOver

```python
reg = linear_model.LinearRegression()
reg.fit(df[['Prev_Close','Open','High','Low','Last','Close','VWAP','Volume','percentage_Deliverble','unemployment_rate','GDP_in_Billion','Per_Capita_income_in_rupees','percentage_increase_in_population','percentage_urban_population','percentage_rural_population']],df.Turnover)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[5436.23,5437,7382,8421,1623,2431,2531,1245,0.6367,5.89,647.83,6748,2.45,3.12,1.22]])
```

Predicting Average Stock Price

```python
reg = linear_model.LinearRegression()
reg.fit(df[['Prev_Close','Open','High','Low','Last','Close','VWAP','Volume','percentage_Deliverble','unemployment_rate','GDP_in_Billion','Per_Capita_income_in_rupees','percentage_increase_in_population','percentage_urban_population','percentage_rural_population']],df.average_stock_price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[5436.23,5437,7382,8421,1623,2431,2531,1245,0.6367,5.89,647.83,6748,2.45,3.12,1.22]])
```
